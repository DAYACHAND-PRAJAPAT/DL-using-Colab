```python
model = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin.gz',binary=True,limit=500000)
```

```python
model['cricket']
```

```python
vec = model['INR'] - model ['India'] + model['England']
model.most_similar([vec])
```
